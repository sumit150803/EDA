```python
#GPA DISTRIBUTION AMONG APPLICANTS
df['GPA'].plot(kind ='hist')
```

```python
#WHICH PALTFORM IS MOSTLY USED TO SEARCH JOBS
df['Primary_Search_Platform'].value_counts().plot(kind='bar')
```

```python
#COMPARSIONO OF STUDENTS WHO RECIEVED OFFERS AND WHO HAVE NOT
sns.countplot(x='Offer_Received',data=df)
```

```python
#COMPARISION OF APPLICANTS WHO ACCEPTED THE OFFER AND WHO DID NOT
job_offered = df[df['Offer_Received']==1]

sns.countplot(data=job_offered,x='Accepted_Offer')
```

```python
#MAXIMUM JOB OFFERED TO WHICH TIER OF UNIVERSITY
sns.countplot(data = job_offered,x='University_Rating')
```

```python
#COMPARISON OF APPICANTS WHO ACCEPTED THE OFFER AND WHO DID NOT
sns.countplot(data=job_offered,x='Accepted_Offer')
```

```python
#HEATMAP ON NUMERICAL COLUMNS

#select all the numerical column
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
corelation_matrix = df[numerical_cols].corr()

sns.heatmap(corelation_matrix,cmap='coolwarm')
```

```python
#Which Major_Category got the most appications
sns.barplot(data=df,x='Major_Category',y='Accepted_Offer',estimator=sum,hue='Major_Category')
```

```python
#Appication submitted vs offer recieved
sns.boxplot(data=df,y='Applications_Submitted',x='Offer_Received',hue='Offer_Received')
```

```python
#Role relevance vs Accepted offer
offers_only =df[df['Offer_Received']==1]
sns.lineplot(data=offers_only,x='Role_Relevance',y='Accepted_Offer')
```
