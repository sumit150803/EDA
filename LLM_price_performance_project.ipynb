```python
from matplotlib import legend
### TOP 20 MODELS RANKED BY INTELLIGENCE INDEX

#top20 = df.nlargest(20,'aa_intelligence_index')
top20 = df.sort_values(by = 'aa_intelligence_index',ascending= False,na_position='last').head(20)

fig , ax = plt.subplots(figsize =(13,8))
bars =ax.barh(range(len(top20)),top20['aa_coding_index']) # Removed 'label' and 'hue' arguments
ax.set_yticks(range(len(top20))) # Set y-axis ticks
ax.set_yticklabels(top20['model_name']) # Set y-axis tick labels to model names
ax.set_xlabel('AA Coding Index') # Add x-axis label
ax.set_ylabel('Model Name') # Add y-axis label
ax.set_title('Top 20 Models by AA Intelligence Index - Coding Index') # Add title
fig.tight_layout() # Adjust layout for better fit
```

```python
### PRICE VS INTELLIGENCE
scatter_df = df.dropna(subset=['aa_intelligence_index']).copy()

fig,ax = plt.subplots(figsize=(14,9))

TIER_COLORS = {
    'Budget': 'green',
    'Mid': 'blue',
    'Premium': 'orange',
    'Ultra': 'red',
    'Free': 'purple'
}

for tier in ['Budget','Mid','Premium','Ultra','Free']:
  mask = scatter_df['pricing_tier'] ==tier
  if mask.any():
    ax.scatter(scatter_df.loc[mask,'blended_cost_usd_per_1m'],
               scatter_df.loc[mask,'aa_intelligence_index'],
               c=TIER_COLORS.get(tier,'#999'),label =tier)

#TOP 10 BEST VALUES
top10v = scatter_df.dropna(subset=['intelligence_per_dollar']).nlargest(10,'intelligence_per_dollar')
# Define C, assuming it's a dictionary for colors. If you have a specific color palette, please define it accordingly.
C = {'accent': 'red'}

for idx, row in top10v.iterrows(): # Corrected to unpack index and row
  short_name = row['model_name'][:28]
  ax.annotate(short_name,
        xy=(row['blended_cost_usd_per_1m'], row['aa_intelligence_index']),
        xytext=(14, 8), textcoords='offset points', fontsize=6.5, fontweight='bold',
        color=C['accent'],
        arrowprops=dict(arrowstyle='->', color=C['accent'], lw=0.7),
        bbox=dict(boxstyle='round,pad=0.2', fc='white', ec=C['accent'], alpha=0.85))
  ax.scatter(row['blended_cost_usd_per_1m'], row['aa_intelligence_index'], # Corrected indentation
              s=130, facecolors='none', edgecolors=C['accent'], linewidth=1.8, zorder=5)

  ax.set_xscale('log')
  ax.set_xlabel('Blended Cost (USD/1m tokens,log scale)')
  ax.set_ylabel('Intelligence Index')
  ax.set_title("PRICE VS INTELLIGENCE - THE VALUE FRONTIER")
  ax.legend()

  plt.tight_layout()


```

```python
fig ,axes = plt.subplots(1,3,figsize = (18,6))

#Intelligence
data_intel = df[df['aa_intelligence_index'].notna()].copy()
data_intel['Source'] = data_intel['is_open_source'].map({True :'Open',False :'Proprietary'})

sns.violinplot(data = data_intel,x='Source',y='aa_intelligence_index',ax = axes[0],cut=0)
axes[0].set_title('Intelligence Index',pad = 10)
axes[0].set_xlabel('')

#Cost

data_cost = df[df['blended_cost_usd_per_1m'].notna() & (df['blended_cost_usd_per_1m']>0)].copy()
data_cost['Source'] = data_cost['is_open_source'].map({True:'Open',False:'Proprietary'})

sns.violinplot(data = data_cost ,x ='Source' ,y ='blended_cost_usd_per_1m',ax = axes[1],inner ='box' ,cut = 0)
axes[1].set_yscale('log')
axes[1].set_title('Blended Cost(Log Scale)',pad =10)
axes[1].set_xlabel('')
axes[1].set_ylabel('USD per 1M tokens')

#Speed
data_Speed  = df[df['output_tokens_per_second'].notna() & (df['output_tokens_per_second'] > 0)].copy()
data_Speed['Source'] = data_Speed['is_open_source'].map({True:'Open' , False:'Proprietary'})

sns.violinplot(data = data_Speed , x='Source' ,y='output_tokens_per_second',inner ='box',cut = 0)
axes[2].set_title('Output Speed(token/sec)',pad = 10)
axes[2].set_xlabel('')

fig.suptitle('Open Source vs Proprietary - Intelligence,Cost & Speed',fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
### TOP 15 BEST VALUED MODELS ON THE BASIS OF INTELLIGENCE PER DOLLAR

val_df = df.dropna(subset=['intelligence_per_dollar'])
val_df = val_df[val_df['intelligence_per_dollar'] > 0]
top15 = val_df.nlargest(15,'intelligence_per_dollar').sort_values('intelligence_per_dollar')

fig,ax = plt.subplots(figsize=(13,8))
bars = ax.barh(range(len(top15)),top15['intelligence_per_dollar'])

ax.set_yticks(range(len(top15)))
ax.set_yticklabels([n[:42] for n in top15['model_name']])
ax.set_xlabel('Intelligence per Dollar')
ax.set_title('TOP 15 BEST VALUE LLMS')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
### SPEED VS COST
df_priced = df[df['blended_cost_usd_per_1m'].notna() & (df['blended_cost_usd_per_1m'] > 0)].copy()
spd_cost = df_priced.dropna(subset=['output_tokens_per_second','aa_intelligence_index']).copy()

df_speed = spd_cost[spd_cost['output_tokens_per_second']>0]

fig ,ax = plt.subplots(figsize =(14,9))

sizes = (spd_cost['aa_intelligence_index']/spd_cost['aa_intelligence_index'].max())*300+15

ax.scatter(spd_cost['blended_cost_usd_per_1m'],spd_cost['output_tokens_per_second'],s=sizes,edgecolors='white',linewidth=0.6,zorder=3)

#Annotate extreme points
fastest = spd_cost.nlargest(3,'output_tokens_per_second')
cheapest_fast = spd_cost[spd_cost['output_tokens_per_second']>100].nsmallest(3,'blended_cost_usd_per_1m')

highlights = pd.concat([fastest,cheapest_fast]).drop_duplicates()

ax.set_xscale('log')
ax.set_xlabel('Blended Cost',fontweight = 'bold')
ax.set_ylabel('Output Tokens per second',fontweight ='bold')
ax.set_title('Speed vs Cost ')
plt.tight_layout()
plt.show()

```
